## Functions that have to be run once

### Function for minimum and maximum float

```python
def calculate_float_stats(condition1, condition2, count_condition1):
    # Define the float ranges for each condition
    float_ranges = {
        'Factory New': (0.02, 0.05),
        'Minimal Wear': (0.077, 0.113),
        'Field-Tested': (0.16, 0.24)
    }
    # Get the float ranges for the given conditions
    range1 = float_ranges.get(condition1)  # Default to (0, 0) if condition not found
    range2 = float_ranges.get(condition2)  # Default to (0, 0) if condition not found
    
    # Calculate the minimum and maximum average floats
    min_avg_float = (range1[0] * count_condition1 + range2[0] * (10 - count_condition1)) / 10
    max_avg_float = (range1[1] * count_condition1 + range2[1] * (10 - count_condition1)) / 10
    
    return min_avg_float, max_avg_float
```

### Function for the input price

```python
def calculate_input_price(wear1, wear2, count1, rarity, collection1, collection2, input_data):
    # # Get the index of the paint wear in the list
    list_of_wears =["Factory New", "Minimal Wear", "Field-Tested", "Well-Worn", "Battle-Scarred"]
    list_wear_index1 = list_of_wears.index(wear1)
    list_wear_index2 = list_of_wears.index(wear2)

    # Get the price of skins from Collection 1 and Collection 2
    price1 = input_data[collection1][rarity][list_wear_index1]
    price2 = input_data[collection2][rarity][list_wear_index2]
    if price1<=0 or price2<=0:
        return 0

    # Calculate the total input price
    total_input_price = (count1 * price1) + ((10 - count1) * price2)

    return total_input_price
```

### Function for the actual calculation of the output of trade up

```python
import numpy as np
def calculate_statistical_gain(avg_float, count1, input_price, rarity, collection1, collection2, output_data):
    if(input_price==0):
        return None
    next_rarity = get_next_rarity(rarity)
    no_of_output1=len(combined_data[collection1][next_rarity])
    no_of_output2=len(combined_data[collection2][next_rarity])
    results={}
    collections=[collection1,collection2]
    counts=[count1,10-count1]
    for i in range(0,2):
# Iterate through each output skin
        for skin_name in combined_data[collections[i]][next_rarity]:
            # Get the minimum float and float range for the skin
            min_float, float_range = combined_data[collections[i]][next_rarity][skin_name][5],combined_data[collections[i]][next_rarity][skin_name][7]
            #Probability of that skin
            prob= counts[i]/(no_of_output1*count1+(10-count1)*no_of_output2)
            # Calculate the output float value
            output_float = min_float + avg_float * float_range
            # Determine the output paint wear based on the output float value
            output_paint_wear = get_paint_wear(output_float)
            # Get the index of the output paint wear
            output_paint_wear_index = ["Factory New", "Minimal Wear", "Field-Tested", "Well-Worn", "Battle-Scarred"].index(output_paint_wear)
            # Get the output price based on the paint wear
            output_price = combined_data[collections[i]][next_rarity][skin_name][output_paint_wear_index]

            # Store the results for this skin
            results[skin_name] = {
                "probability": prob,
                "output_float": output_float,
                "output_price": output_price,
            }
    # Calculate the expected output price (before tax)
    expected_output_price = sum(data["probability"] * data["output_price"] for data in results.values())
    
    # Adjust the expected output price for Steam tax (100/115)
    adjusted_expected_output_price = expected_output_price * (100 / 115)
    
    # Calculate the statistical gain ratio
    statistical_gain_ratio = adjusted_expected_output_price / input_price if input_price > 0 else 0
    #print(adjusted_expected_output_price,input_price,statistical_gain_ratio)
    return {
        "avg_float": avg_float,
        "input_price": input_price,
        "adjusted_expected_output_price": adjusted_expected_output_price,
        "statistical_gain_ratio": statistical_gain_ratio,
        "results": results
    }
```

### Side Functions

```python
def get_paint_wear(float_value):
    if float_value < 0.07:
        return "Factory New"
    elif float_value < 0.15:
        return "Minimal Wear"
    elif float_value < 0.38:
        return "Field-Tested"
    elif float_value < 0.45:
        return "Well-Worn"
    else:
        return "Battle-Scarred"

# Define the get_next_rarity function
def get_next_rarity(rarity):
    rarity_order = ["Consumer", "Industrial", "Mil-Spec", "Restricted", "Classified", "Covert"]
    try:
        index = rarity_order.index(rarity)
        if index + 1 < len(rarity_order):
            return rarity_order[index + 1]
        else:
            return None
    except ValueError:
        return None
```
